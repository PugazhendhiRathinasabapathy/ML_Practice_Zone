```python
from sklearn.datasets import fetch_california_housing
```

```python
import numpy as np 
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
dataset = fetch_california_housing()
```

```python
dataset
```

```python
type(dataset)
```

```python
df = pd.DataFrame(dataset.data)
df.head()
```

```python
type(df)
```

```python
df.columns = dataset.feature_names
df.head()
```

target variable is the median house value for California districts,\nexpressed in hundreds of thousands of dollars ($100,000)

```python
df['Price'] = dataset.target
```

```python
df
```

```python
# Spliting the features(X) and target(y) 
y = df['Price']
```

```python
X = df.drop(['Price'], axis=1)
```

```python
X
```

```python
from sklearn.linear_model import LinearRegression

mod = LinearRegression()
mod.fit(X,y)
```

```python
mod.predict(pd.DataFrame(np.array([[8.3014,21.0,6.238137,0.971880,2401.0,2.109842,37.86,-122.22]])))
```

```python
y
```

```python
pred=mod.predict(X)
```
