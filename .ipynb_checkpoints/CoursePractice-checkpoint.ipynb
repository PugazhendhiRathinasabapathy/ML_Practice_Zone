```python
from sklearn.datasets import fetch_california_housing
```

```python
import numpy as np 
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
dataset = fetch_california_housing()
```

```python
dataset
```

```python
type(dataset)
```

```python
df = pd.DataFrame(dataset.data)
df.head()
```

```python
type(df)
```

```python
df.columns = dataset.feature_names
df.head()
```

target variable is the median house value for California districts,\nexpressed in hundreds of thousands of dollars ($100,000)

```python
df['Price'] = dataset.target
```

```python
df
```

```python
print(fetch_california_housing()['DESCR'])
```

```python
# Spliting the features(X) and target(y) 
y = df['Price']
```

```python
X = df.drop(['Price'], axis=1)
```

```python
X
```

```python
from sklearn.linear_model import LinearRegression

mod = LinearRegression()
mod.fit(X,y)
```

```python
mod.predict(pd.DataFrame(np.array([[8.3014,21.0,6.238137,0.971880,2401.0,2.109842,37.86,-122.22]])))
```

```python
y
```

```python
pred=mod.predict(X)
```

```python
plt.scatter(pred,y)
```

```python
#Getting familiar with the metrics.
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score

mse = mean_squared_error(y,pred) #mean sqaured error
print("MSE = ",mse)

rmse = np.sqrt(mse) #root mean squared error
print("RMSE = ",rmse)

mae = mean_absolute_error(y,pred) #mean absolute error
print("MAE = ",mae)

r2 = r2_score(y,pred) #R-sqaure error
print("R2 = ",r2)
```

## Ridge Regression 

```python
# Train test split 
from sklearn.model_selection import train_test_split

X_train,X_test,y_train,y_test = train_test_split(X, y ,test_size=0.2, random_state=41) #80-20 split -> test_size=0.2
```

```python
X_train
```

```python
from sklearn.linear_model import Ridge

rid = Ridge(alpha=0.1)
rid.fit(X_train,y_train)
```

```python
y_pred=rid.predict(X_test)
```

```python
r2_ridge = r2_score(y_test,y_pred)
print(r2_ridge)
```

```python
print(r2," = ", r2_ridge)
```

```python
#Hyperparameter tuning by changing the alpha value.
tuned_ridge = Ridge(alpha=100)
tuned_ridge.fit(X_train,y_train)
```

```python
y_pred2 =  tuned_ridge.predict(X_test)
```

```python
r2_tuned_ridge = r2_score(y_test,y_pred2)
print(r2_ridge," -> ",r2_tuned_ridge)
```


We have tuned the hyperparameters but the model performance hasn't improved. We might have got the wrong alpha value


How to get the correct alpha value?

Using cross-validation techniques like GridSearchCV 

```python
from sklearn.model_selection import GridSearchCV

ridge = Ridge()

parameter_alpha = {'alpha': np.logspace(-5,5,10)}
```

```python
parameter_alpha
```

```python
#Setting up the model of GridSearchCVabs
ridge_reg = GridSearchCV(estimator=ridge,param_grid=parameter_alpha,scoring='r2',cv=5)

#Building model
ridge_reg.fit(X_train,y_train)
```

```python
# Print best parameter value and its corresponding score 
print(ridge_reg.best_params_)
print(ridge_reg.best_score_)
```

## Lasso regression L1 regularization 

```python
from sklearn.linear_model import Lasso

lasso = Lasso()

lasso_reg =  GridSearchCV(lasso,parameter_alpha,scoring='r2',cv=5)
lasso_reg.fit(X_train,y_train)
```

```python
print(lasso_reg.best_params_)
print(lasso_reg.best_score_)
```

```python
y_pred_ridge = ridge_reg.predict(X_test)
y_pred_lasso = lasso_reg.predict(X_test)
```

```python
plt.figure(figsize=(8, 5))
plt.scatter(y_test, y_pred_ridge, alpha=0.5, color='blue', label="Ridge Predictions")
plt.scatter(y_test, y_pred_lasso, alpha=0.5, color='green', label="Lasso Predictions")
plt.plot([min(y_test), max(y_test)], [min(y_test), max(y_test)], color='red', linestyle='dashed', label="Ideal Fit")  
plt.xlabel("Actual Values (y_test)")
plt.ylabel("Predicted Values")
plt.title("Actual vs. Predicted: Ridge vs. Lasso")
plt.legend()
plt.show()
```
